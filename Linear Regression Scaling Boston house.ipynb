## Import Libraries

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

from statsmodels.stats.outliers_influence import variance_inflation_factor

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
```

```python
warnings.filterwarnings("ignore")
```

## Data Gathering

```python
from sklearn.datasets import load_boston
```

```python
data_dict = load_boston()
```

```python
df = pd.DataFrame(data = data_dict.data, columns=data_dict.feature_names)
```

```python
df
```

```python
df["House_Price"] = data_dict.target
```

## Feature Engineering

## Train The Model

```python
x = df.drop("House_Price", axis = 1)
y = df["House_Price"]
```

```python
x.head()
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(x_train,y_train)
```

```python
y_pred = lr_model.predict(x_test)
```

## Evaluation of model

```python
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test,y_pred)
r2score = r2_score(y_test,y_pred)

print(f"Mean Squared Error: {mse:0.4f}")
print(f"Root mean Squared Error: {rmse:0.4f}")
print(f"Mean Absolute Error: {mae:0.4f}")
print(f"R2 Score: {r2score:0.4f}")
```

## Scaling the data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(x)

x_scaled = scaler.transform(x)

x_scaled
```

```python
x_scaled = pd.DataFrame(x_scaled, columns = x.columns)
```

## Training Model After Scaling The Data

```python
xs_train, xs_test, y_train, y_test = train_test_split(x_scaled,y)
```

```python
lr_model = LinearRegression()
lr_model.fit(xs_train,y_train)
y_pred = lr_model.predict(xs_test)
```

```python
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test,y_pred)
r2score = r2_score(y_test,y_pred)

print(f"Mean Squared Error: {mse:0.4f}")
print(f"Root mean Squared Error: {rmse:0.4f}")
print(f"Mean Absolute Error: {mae:0.4f}")
print(f"R2 Score: {r2score:0.4f}")
```
